Numpy Arrays

```python
list1=[1,2,3,4]
print(list1)
print(type(list1))
```

```python
#incase of numpy not seprated by comma
#nd=n-dimensional similar to matrix
np_array=np.array([1,2,3,4])
print(np_array)
print(type(np_array))
```

creating 1d arrays

```python
np_array=np.array([1,2,3,4])
print(np_array)
np_array.shape
b=np.array([(1,2,3,4),(5,6,7,8)])
print(b)
b.shape
c=np.array([(1,2,3,4),(4,5,6,7)],dtype=float)
print(c)
#default datatype is int if u dont mention
```

initial placeholders

```python
x=np.zeros((4,5))
print(x)
#create np values of 4 rows 5 colums with all zeros
# np array ones
y=np.ones((2,3))
print(y)
#array of a particular value
z=np.full((3,3),90)
print(z)
#create identity matrix
a=np.eye(4)# row=col for identity
print(a)
```

```python
#numpy array with random values
b=np.random.random((3,4))
print(b)
#b/w 0 and 1
```

```python
#random integers array
c=np.random.randint(10,50,(3,5))
print(c)
```

```python
#array of evenly spaced values..also can specify no. of values required
d=np.linspace(10,50,5)
print(d)
```

```python
#specyfying step no.
e=np.arange(10,50,5)
#5= difference
print(e)
```

```python
#covert list to numpy arraay
list2=[10,10,30,40,50]
np_Arr=np.asarray(list2)
print(np_Arr)
type(np_Arr)
```

```python
arr1=np.random.randint(10,50,(4,4))
print(arr1)
print(arr1.shape)# return no. of rows and column
print(arr1.ndim)#return dimensions(2d array)
print(arr1.size)#no.of elements
print(arr1.dtype)#datatype-64 bit
```
