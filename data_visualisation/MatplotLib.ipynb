```python
#numpy for data ,matplot lib  #useful for making plots

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
z=np.log(x)
#log function
plt.plot(x,z)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('log(x)')
plt.title('log graph')
plt.show()
#sine function
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Sine wave')
plt.grid(True)
plt.show()
```

```python
#plotting a parabola
a=np.linspace(-10,10,20)
b=a*a
plt.plot(a,b)
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('parabola')
plt.show()
```

```python
#using colors
plt.plot(a,b,'g.')
```

```python
plt.plot(a,b,'x')
plt.show()
```

```python
c=np.linspace(-10,10,50)
plt.plot(c,np.sin(c),'b--')
plt.plot(c,np.cos(c),'r--')
plt.show()
```

bar graphs


```python
fig=plt.figure()
axs=fig.add_axes([0,0,1,1])
books=['english','cse','french','maths','python']#lists
students=[100,35,67,33,67]
axs.bar(books, students) # Add this line to plot the bars
axs.set_xlabel('BOOKS')
axs.set_ylabel('STUDENTS')
plt.show()
```

PieCharts

```python
fig1=plt.figure()
ax=fig1.add_axes([0,0,1,1])
ax.pie(students,labels=books,autopct='%1.2f%%')
#autopct is to tell how many floating point we need like 1.2 or 1.22 or 1.222
plt.show()
```

Scattered Plot

```python
x=np.linspace(0,10,30)
y=np.sin(x)
z=np.cos(x)
fig2=plt.figure()
ax=fig2.add_axes([0,0,1,1])
ax.scatter(x,y,color='r')
ax.scatter(x,z,color='b')
plt.show()
```

3-D Plot

```python
fig3=plt.figure()
ax=plt.axes(projection='3d')
z=20*np.random.random(100)
x=np.sin(z)
y=np.cos(z)
ax.scatter(x,y,z,c=z,cmap='gist_rainbow_r')
plt.show()
```
